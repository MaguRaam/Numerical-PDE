```python
from mpl_toolkits.mplot3d import Axes3D
import numpy
from matplotlib import pyplot, cm
%matplotlib inline

```

```python
b = numpy.loadtxt('b.dat')

def plot_2D(x,y,p):
    X, Y = numpy.meshgrid(x, y)
    fig = pyplot.figure(figsize=(11, 7), dpi=100)
    pyplot.contourf(X, Y, p, alpha=0.5, cmap=cm.viridis)
    pyplot.colorbar()
    pyplot.contour(X, Y, p, cmap=cm.viridis)
    pyplot.xlabel('X')
    pyplot.ylabel('Y')
    
```

## Solve poisson equation

$\nabla^{2}u = -2n^{2}\pi^{2}\sin(n\pi x)\sin(n\pi y)$  


## Right Handside b

```python
x = numpy.loadtxt('x.dat')
y = numpy.loadtxt('y.dat')
b = numpy.loadtxt('b.dat')

plot_2D(x,y,b)
```

```python
u = numpy.loadtxt('u.dat')

plot_2D(x,y,u)
```

## uexact
sin(npix) sin(npiy)

```python
uexact = numpy.loadtxt('uexact.dat')

plot_2D(x,y,uexact)
```
