```python
## Project b on eigenvector and get b~dfrom mpl_toolkits.mplot3d import Axes3D
import numpy
from matplotlib import pyplot, cm
%matplotlib inline
from math import log

```

```python
N  = numpy.array([64,128,256,512])
l2 = numpy.array([0.0210351,0.00512903,0.00127439,0.000318109]) 
```

```python
pyplot.loglog(N, l2)
pyplot.xlabel("N")
pyplot.ylabel("L2 error")
```

```python

for i in range(1,numpy.size(l2)):
    covergence_rate = (log(l2[i-1]) - log(l2[i]))/log(2)
    print(covergence_rate)
```
