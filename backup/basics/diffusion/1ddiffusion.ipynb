```python
import numpy
from matplotlib import pyplot 
```

```python
#grid:
L=2.0
N=500
h = L/(N-1)

#stability:
sigma = 0.5

#diffusion coefficient:
mu = 0.9

#time 
T = 0.01
dt = sigma*h*h/mu 
```

```python
#initialize:
u = numpy.ones(N)
u[int(0.5/h):int(1.0/h)] = 2.0

#plot IC:
pyplot.plot(numpy.linspace(0,L,N),u)
```

```python
#start time loop:
t = 0.0
while t<T:
    t+=dt
    un = u.copy()
    for i in range(1,N-1):
        u[i] = un[i] + sigma*(u[i+1] - 2*u[i] + u[i-1])

#plot the final solution 
x = numpy.linspace(0,L,N)
pyplot.plot(x,u)



```
