```python
from mpl_toolkits.mplot3d import Axes3D
import numpy
from matplotlib import pyplot, cm
%matplotlib inline
from math import log
```

```python
def plot_2D(x,y,p):
    X, Y = numpy.meshgrid(x, y)
    fig = pyplot.figure(figsize=(11, 7), dpi=100)
    pyplot.contourf(X, Y, p, alpha=0.5, cmap=cm.viridis)
    pyplot.colorbar()
    pyplot.contour(X, Y, p, cmap=cm.viridis)
    pyplot.xlabel('X')
    pyplot.ylabel('Y')
```

```python
x = numpy.loadtxt('x.dat')
y = numpy.loadtxt('y.dat')
b = numpy.loadtxt('U.dat')

plot_2D(x,y,b)
```

```python
x = numpy.loadtxt('x.dat')
y = numpy.loadtxt('y.dat')
b = numpy.loadtxt('Uexact.dat')

plot_2D(x,y,b)
```

## L2 error

```python
N  = numpy.array([64,128,256,512])
l2 = numpy.array([0.0210351,0.00512903,0.00127439,0.000318109])
```

```python

pyplot.loglog(N, l2)
pyplot.xlabel("N")
pyplot.ylabel("L2 error")
```

```python
#Rate of convergence:
for i in range(1,numpy.size(l2)):
    covergence_rate = (log(l2[i-1]) - log(l2[i]))/log(2)
    print(covergence_rate)
```

## Linfty error

```python
linfty = numpy.array([0.0210351,0.00512903,0.00127439,0.000318109])


pyplot.loglog(N, linfty)
pyplot.xlabel("N")
pyplot.ylabel("Linfty error")
```

```python
#Rate of convergence:
for i in range(1,numpy.size(linfty)):
    covergence_rate = (log(linfty[i-1]) - log(linfty[i]))/log(2)
    print(covergence_rate)
```
