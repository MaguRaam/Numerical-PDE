```python
from mpl_toolkits.mplot3d import Axes3D
import numpy
from matplotlib import pyplot, cm
%matplotlib inline

```

```python
def plot_2D(x,y,p):
    X, Y = numpy.meshgrid(x, y)
    fig = pyplot.figure(figsize=(11, 7), dpi=100)
    pyplot.contourf(X, Y, p, alpha=0.5, cmap=cm.viridis)
    pyplot.colorbar()
    pyplot.contour(X, Y, p, cmap=cm.viridis)
    pyplot.xlabel('X')
    pyplot.ylabel('Y')
    
```

## Right Handside b

```python
x = numpy.loadtxt('plot/x.dat')
y = numpy.loadtxt('plot/y.dat')
b = numpy.loadtxt('plot/b.dat')

plot_2D(x,y,b)
```

## Components of Right handside in eigenbasis b~

```python
b_tilde = numpy.loadtxt('plot/b_tilde.dat')

plot_2D(x,y,b_tilde)
```

## Solution in eigen basis x~

```python

u_tilde = numpy.loadtxt('plot/u_tilde.dat')
plot_2D(x,y,u_tilde)
```

## Solution in standard basis x

```python
u = numpy.loadtxt('plot/u.dat')
plot_2D(x,y,u)
```
