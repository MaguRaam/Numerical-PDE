```python
xc = numpy.loadtxt('plot/x.dat')
yc = numpy.loadtxt('plot/y.dat')
rho = numpy.loadtxt('plot/u.dat')
                    
X, Y = numpy.meshgrid(xc, yc)      
fig = pyplot.figure(figsize=(6,6), dpi=100)
# plotting the pressure field as a contour
pyplot.contourf(X, Y, rho, alpha=0.5, cmap=cm.viridis)  
pyplot.colorbar()
# plotting the density field outlines
pyplot.contour(X, Y, rho, cmap=cm.viridis)                        
 
```
