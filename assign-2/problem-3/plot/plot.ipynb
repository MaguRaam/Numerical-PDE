```python
## Project b on eigenvector and get b~dfrom mpl_toolkits.mplot3d import Axes3D
import numpy
from matplotlib import pyplot, cm
%matplotlib inline

```

```python


def plot_2D(x,y,p):
    X, Y = numpy.meshgrid(x, y)
    fig = pyplot.figure(figsize=(11, 7), dpi=100)
    pyplot.contourf(X, Y, p, alpha=0.5, cmap=cm.viridis)
    pyplot.colorbar()
    pyplot.contour(X, Y, p, cmap=cm.viridis)
    pyplot.xlabel('X')
    pyplot.ylabel('Y')
    
```

## Solve poisson equation

$\nabla^{2}u = -2n^{2}\pi^{2}\sin(n\pi x)\sin(n\pi y)$  


## Right Handside b

```python
x = numpy.loadtxt('x.dat')
y = numpy.loadtxt('y.dat')
b = numpy.loadtxt('b.dat')

plot_2D(x,y,b)
```

## Project b on eigenvector and get b~

```python
btilde = numpy.loadtxt('btilde.dat')

plot_2D(x,y,btilde)
```

## Solve for u~ = b~/$\lambda$

```python
utilde = numpy.loadtxt('utilde.dat')

plot_2D(x,y,utilde)
```

## Expan u~ using eigen basis and get u

```python
u = numpy.loadtxt('u.dat')

plot_2D(x,y,u)
```
