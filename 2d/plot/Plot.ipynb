```python
from mpl_toolkits.mplot3d import Axes3D
import numpy
from matplotlib import pyplot, cm
%matplotlib inline
from math import log
```

```python
def plot_2D(x,y,p):
    X, Y = numpy.meshgrid(x, y)
    fig = pyplot.figure(figsize=(11, 7), dpi=100)
    pyplot.contourf(X, Y, p, alpha=0.5, cmap=cm.viridis)
    pyplot.colorbar()
    pyplot.contour(X, Y, p, cmap=cm.viridis)
    pyplot.xlabel('X')
    pyplot.ylabel('Y')
```

```python
x = numpy.loadtxt('x.dat')
y = numpy.loadtxt('y.dat')
u = numpy.loadtxt('U.dat')

plot_2D(x,y,u)
```

```python
x = numpy.loadtxt('x.dat')
y = numpy.loadtxt('y.dat')
uexact = numpy.loadtxt('Uexact.dat')

plot_2D(x,y,uexact)
```

```python
x = numpy.loadtxt('x.dat')
y = numpy.loadtxt('y.dat')
b = numpy.loadtxt('B.dat')

plot_2D(x,y,b)
```

```python
N  = numpy.array([16,32,64,128])
l2_space = numpy.array([1.3137e-05,3.37235e-06,9.63036e-07,3.62678e-07])
linfty_space = numpy.array([2.6274e-05,6.74471e-06,1.92607e-06,7.25356e-07])

fig,ax = pyplot.subplots()

pyplot.loglog(N, l2_space,'-b',label = 'L2 error',marker = 'o')
pyplot.loglog(N, linfty_space,'--r',label = 'Linfty error',marker = 'o')
leg = ax.legend()
pyplot.title("Convergence in space")
pyplot.xlabel("N")
pyplot.ylabel("Error")
```

```python
dt  = numpy.array([16,32,64,128])
l2_time = numpy.array([1.46055e-05,1.33778e-05,1.3072e-05, 1.29956e-05])
linfty_time = numpy.array([2.9211e-05,2.67556e-05,2.61439e-05,2.59912e-05])

fig,ax = pyplot.subplots()

pyplot.loglog(dt, l2_time,'-b',label = 'L2 error',marker = 'o')
pyplot.loglog(dt, linfty_time,'--r',label = 'Linfty error',marker = 'o')
leg = ax.legend()
pyplot.title("Convergence in time")
pyplot.xlabel("dt")
pyplot.ylabel("Error")
```

```python
#Rate of convergence:
for i in range(1,numpy.size(l2_time)):
    covergence_rate = (log(l2_time[i-1]) - log(l2_time[i]))/log(2)
    print(covergence_rate)
```
