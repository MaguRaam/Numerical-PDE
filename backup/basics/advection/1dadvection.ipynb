```python
import numpy as numpy
from matplotlib import pyplot
 
```

```python
L = 2.0
nx = 30
dx = L/(nx-1)

c  = 1

cfl =1
T = 0.5
dt = cfl*dx/c
```

```python
x = numpy.linspace(0.0,2.0,nx)
```

```python
 u = numpy.ones(nx)
 u[int(0.5/dx) : int(1/dx)] = 2.0
```

```python
pyplot.plot(x,u)
```

```python
t = 0.0
while t<T:
    t+=dt
    un = u.copy()
    for i in range(1,nx-1):
        u[i] = un[i] - (c*dt/dx)*(un[i] - un[i-1])
        #u[i] = un[i] - ((c*dt/2*dx)*(un[i+1] - un[i-1]))
   
         
   
```

```python
pyplot.plot(x,u)
```
